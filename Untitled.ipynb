```python
alpha_channel=alpha[:,:,0]
r_channel, g_channel, b_channel = cv2.split(img)
img_rgba = cv2.merge((r_channel, g_channel, b_channel, alpha_channel))
plt.imshow(img_rgba) ##opencv 显示时rgba值可以是0-1之间，但若直接保存到磁盘，就会变成黑色
```

```python
img_rgba=cv2.cvtColor(img_rgba, cv2.COLOR_RGBA2BGRA) ##opencv默认采用bgr或bgra作为读取和保存的格式，因此需要转换
cv2.imwrite("img_rgba.png", img_rgba) 
```

```python
alpha_channel=alpha[:,:,0]
r_channel, g_channel, b_channel = cv2.split(img)
r_channel=r_channel/255
g_channel=g_channel/255
b_channel=b_channel/255
alpha_channel=alpha_channel/255
alpha_channel
```

```python
bgcolor=[1,1,1] #背景色是白色

r_channel=(1-alpha_channel)*bgcolor[0]+alpha_channel*r_channel
g_channel=(1-alpha_channel)*bgcolor[1]+alpha_channel*g_channel
b_channel=(1-alpha_channel)*bgcolor[2]+alpha_channel*b_channel

img_rgb_bg = cv2.merge((r_channel, g_channel, b_channel))
plt.imshow(img_rgb_bg)
```

```python
img_rgb_bg=(img_rgb_bg*255).astype(np.uint8) ##把rgba值转换回0-255的整数
img_rgb_bg[img_rgb_bg>255]=255
img_rgb_bg=cv2.cvtColor(img_rgb_bg, cv2.COLOR_RGB2BGR)
cv2.imwrite("img_rgb_bg.png",  img_rgb_bg)
```
